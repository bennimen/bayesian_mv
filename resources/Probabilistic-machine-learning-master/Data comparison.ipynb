```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# R

```python
with open("2.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile)
    T_R = [col['v3v4diff_tot_pps_r_cap_0'] for col in reader]
    T_R = list(map(float, T_R))
    T_R = np.array(T_R)
with open("2.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile)
    x_R = [col['v3v4diff_x_pps_r_cap_0'] for col in reader]
    x_R = list(map(float, x_R))
    x_R = np.array(x_R)
with open("2.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile)
    y_R = [col['v3v4diff_y_pps_r_cap_0'] for col in reader]
    y_R = list(map(float, y_R))
    y_R = np.array(y_R)


print ("Total_R max :", np.max(T_R))
print ("Total_R min :", np.min(T_R))
print ("Total_R mean:", np.mean(T_R))
print ("Total_R std :", np.std(T_R))


fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
theta = np.arctan(y_R/x_R)
theta = 2 * np.pi * theta
c = ax.scatter(theta, T_R, s=1)
plt.title('R displacement')
plt.savefig('R Circle.png')

fig = plt.figure()
ax = fig.add_subplot(111)
plt.title('R displacement')
plt.boxplot(T_R)
plt.savefig('R Boxplot.png')
```
